```python
champs_win = {}
champs_showup = {}
```

```python
# Iterate through each row of the DataFrame
for index, row in m_cluster.iterrows():
    # Process red champions
    for champs in row['red_champions']:
        champs_showup[champs] = champs_showup.get(champs, 0) + 1
        if row['red_win'] > 0.5:
            champs_win[champs] = champs_win.get(champs, 0) + 1
    
    # Process blue champions
    for champs in row['blue_champions']:
        champs_showup[champs] = champs_showup.get(champs, 0) + 1
        if row['blue_win'] > 0.5:
            champs_win[champs] = champs_win.get(champs, 0) + 1
```

```python
import numpy as np
def winrate(row):
    championid = row['championid']
    win =  champs_win.get(championid, 0)
    showup = champs_showup.get(championid, 0)
    return win/showup if showup != 0 else np.nan
```

```python
c_winrate['winrate_x'] = c_winrate.apply(
    lambda row: winrate(row),
    axis=1
)
```

```python
duo_showup = {}
duo_win = {}
```

```python
import itertools

# Iterate through each row of the DataFrame
for index, row in m_cluster.iterrows():
    # Process red champions
    reds = list(row['red_champions'])
    combinations = list(itertools.combinations(reds,2))
    sorted_combinations = [tuple(sorted(combo)) for combo in combinations]
    for combo in sorted_combinations:
        duo_showup[combo] = duo_showup.get(combo, 0) + 1
        if row['red_win'] > 0.5:
            duo_win[combo] = duo_win.get(combo, 0) + 1
    # Process blue champions
    blues = list(row['blue_champions'])
    combinations = list(itertools.combinations(blues, 2))
    sorted_combinations = [tuple(sorted(combo)) for combo in combinations]
    
    for combo in sorted_combinations:
        duo_showup[combo] = duo_showup.get(combo, 0) + 1
        if row['blue_win'] > 0.5:
            duo_win[combo] = duo_win.get(combo, 0) + 1
```
